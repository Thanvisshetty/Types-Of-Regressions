# TYPES OF REGRESSION (7 TYPES)

## 1. Linear Regression
Used When: There's a linear relationship between one independent variable and one dependent variable.

Explanation:
Linear Regression fits a straight line (y = mx + c) through the data points that best predicts the target variable.

Example: Predicting salary based on years of experience.

Key Point:
Assumes data is linearly correlated.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([30000, 35000, 50000, 55000, 60000])

# Model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Output
print("Predicted salaries:", y_pred)

# Plot
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title("Simple Linear Regression")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## 2. Multiple Linear Regression
Used When: More than one independent variable influences the target.

Explanation:
It extends linear regression by fitting a plane/hyperplane in multidimensional space. The equation is:
y = b0 + b1*x1 + b2*x2 + ... + bn*xn

Example: Predicting house prices based on size, number of rooms, and age.

Key Point:
Each feature contributes its own weight to the prediction.

```python
# Data: [area, bedrooms, age]
X = np.array([[1000, 2, 5], [1500, 3, 10], [2000, 4, 15]])
y = np.array([200000, 250000, 300000])

# Model
mlr = LinearRegression()
mlr.fit(X, y)
y_pred = mlr.predict(X)

# Plot predicted vs actual
plt.scatter(y_pred, y, color='green')
plt.title("2. Multiple Linear Regression: Predicted vs Actual")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## 6. Logistic Regression
Used When: The target variable is categorical (e.g., 0 or 1, True or False).

Explanation:
Despite its name, logistic regression is used for classification. It uses the sigmoid function to predict probabilities.

Example: Predicting if a student passes or fails based on their marks.

Key Point:
Output is a probability (between 0 and 1), then thresholded to get a class.

```python
# Data
X = np.array([[30], [50], [60], [80], [90]])
y = np.array([0, 0, 1, 1, 1])

# Model
log_reg = LogisticRegression()
log_reg.fit(X, y)

# Predict probability
X_range = np.linspace(30, 100, 300).reshape(-1, 1)
probs = log_reg.predict_proba(X_range)[:, 1]

# Plot
plt.scatter(X, y, color='blue', label="Data")
plt.plot(X_range, probs, color='red', label="Sigmoid Curve")
plt.title("6. Logistic Regression")
plt.xlabel("Marks")
plt.ylabel("Pass Probability")
plt.legend()
plt.show()
```

## 7. ElasticNet Regression
Used When: You want a balance between Ridge and Lasso.

Explanation:
It combines L1 and L2 regularization:
Loss = RSS +  * (l1_ratio * |coef| + (1 - l1_ratio) * coef)

Example: Useful when dataset has many features, some of which are correlated and some irrelevant.

Key Point:
Offers both feature selection and coefficient shrinkage.

```python
# Data
X = np.array([[1, 2], [2, 3], [3, 4]])
y = np.array([3, 5, 7])

# Model
enet = ElasticNet(alpha=0.1, l1_ratio=0.5)
enet.fit(X, y)
y_pred = enet.predict(X)

# Plot
plt.scatter(y_pred, y, color='brown')
plt.title("7. ElasticNet Regression: Predicted vs Actual")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
